```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
np.random.seed(17)
tf.random.set_seed(17)
```

```python
df = pd.read_csv('prepared_posts_df.csv')
```

```python
# создадим матрицы признаков и ответов
X = df.drop(columns=['rating']).to_numpy()
Y = df['rating'].to_numpy()
```

```python
# разделим датасет на части: тренировочную, валидационную и тестовую (в пропорции 70/15/15)
# это можно сделать как специальными функциями в фреймворках, так и обычными слайсами

val_frac = 0.15
test_frac = 0.15

x_train = X[:round((1 - val_frac - test_frac) * len(X))]
x_val = X[round((1 - val_frac - test_frac) * len(X)):round((1 - test_frac) * len(X))]
x_test = X[round((1 - test_frac) * len(X)):]

y_train = Y[:round((1 - val_frac - test_frac) * len(Y))]
y_val = Y[round((1 - val_frac - test_frac) * len(Y)):round((1 - test_frac) * len(Y))]
y_test = Y[round((1 - test_frac) * len(Y)):]
```

```python
'''
Задание 1.  Создайте полносвязную нейросеть для нашей задачи предсказания рейтинга поста 
с тремя промежуточными слоями по 100 нейронов и функцией активации ReLU. 
Инициализируйте её весами из файла homework_init_weights.h5. 
Какое значение mean_absolute_error получится, 
если в таком состоянии прогнать нейросеть на всём датасете? Ответ округлите до целого.
'''
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(100, activation='relu', input_shape=(X.shape[1],)),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(1, activation=tf.keras.activations.linear)
])
model.load_weights('homework_init_weights_DZ.h5')
```

```python
model.compile(
    loss=tf.keras.losses.MSE,
    optimizer=tf.keras.optimizers.Adam(),
    metrics=[tf.keras.metrics.MeanAbsoluteError()]
)
```

```python
model.evaluate(X, Y)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(100, activation='relu', input_shape=(X.shape[1],)),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(1, activation=tf.keras.activations.linear)
])
model.load_weights('homework_init_weights_DZ.h5')
```

```python

model.compile(
    loss=tf.keras.losses.MSE,

    optimizer=tf.keras.optimizers.Adam(),

    metrics=[tf.keras.metrics.MeanAbsoluteError(),
             tf.keras.metrics.RootMeanSquaredError()]
)
```
